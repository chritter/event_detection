### Combine and write out data

```python
sentences = data['holdin_instances'] + data['holdout_instances']
labels = data['holdin_labels'] + data['holdout_labels']
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'sentences':sentences,'label':labels})
```

```python
df.head()
```

```python
import os
title=[]
file_content=[]
date=[]
file_id=[]
for i,file in enumerate(os.listdir("osfstorage-archive/bratannotationfiles")):
    if file.endswith(".txt"):
        file_name=os.path.join("osfstorage-archive/bratannotationfiles", file)
        ftmp = file_name[:-4].split('/')[-1]
        title.append((' '.join(ftmp.split('_')[:-1]).strip()))
        date.append(ftmp.split('_')[-1])
        file_id.append(i)
        with open(file_name) as f:
            lines = f.readlines()
            lines = ' '.join(lines)
            file_content.append(lines.replace('\n',''))
```

```python
df_full_corpus=pd.DataFrame({'file_id':file_id,'title':title,'date':date,'text':file_content})
```

```python
df_full_corpus.head()
```

```python
def find_doc(sentence):
    for i,c in enumerate(file_content):
        if sentence in c:
            return title[i]
    print('counld not find document')
    return ''
def find_doc2(sentence):
    for i,c in enumerate(file_content):
        if sentence in c:
            return date[i]
    print('counld not find document')
    return ''
def find_doc3(sentence):
    for i,c in enumerate(file_content):
        if sentence in c:
            return file_id[i]
    print('counld not find document')
    return ''
```

```python
df['title'] = df['sentences'].apply(find_doc)
df['publication_date'] = df['sentences'].apply(find_doc2)
df['file_id'] = df['sentences'].apply(find_doc3)
```

```python
df.head()
```

```python
list_unique_labels = list(data['meta_stats_alldata']['label_count'].keys())
```

```python
for label1 in list_unique_labels:
    df[label1] = 0
```

```python
def check_labels(row):
    llabels = row['label']
    if type(llabels) == list:
        for lel in llabels:
            row[lel] = 1
    else:
        if llabels == -1:
            row['-1'] = 1
    return row
```

```python
df_new = df.apply(check_labels,axis=1)
```

### Assign labels to documents

```python
doc_ids = df_full_corpus_body['file_id'].values
labels=[]
for i,id1 in enumerate(doc_ids):
    labels1=[]
    for id2,label in df_new_subset:
        if id1 == id2:
            labels1+=label
            #print('found id')
    labels.append(labels1)
```

```python
df_full_corpus_body['labels'] = labels
```

### Map labels into multiple single features; each feature appearing once

```python
list_unique_labels = ['-1',
 'Profit',
 'Dividend',
 'MergerAcquisition',
 'SalesVolume',
 'BuyRating',
 'QuarterlyResults',
 'TargetPrice',
 'ShareRepurchase',
 'Turnover',
 'Debt']
```

```python
for label1 in list_unique_labels:
    df_full_corpus_body[label1] = 0
```

```python
def check_labels(row):
    llabels = row['labels']
    #if type(llabels) == list:
    if len(llabels)>0:
        for lel in llabels:
            row[lel] = 1
    else:
            row['-1'] = 1
    return row
```

```python
df_full_corpus_body = df_full_corpus_body.apply(check_labels,axis=1)
```
